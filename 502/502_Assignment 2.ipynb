```python
import pandas as pd
from matplotlib import pyplot as plt

data = pd.read_csv('insurance.csv')
data.head()
```

```python
plt.scatter(data['bmi'], data['charges'],c = data['smoker'], alpha = 0.7)
plt.xlabel('BMI')
plt.ylabel('Charges')
cbar = plt.colorbar(ticks=[0, 1]) # add color bar with specific ticks
cbar.set_ticklabels(['Non-Smoker','Smoker'])
plt.show()
```

```python
smoker = data.loc[data['smoker'] == 1, 'charges']
nonsmoker = data.loc[data['smoker'] == 0, 'charges']

plt.hist(smoker,12)
plt.hist(nonsmoker,12,alpha=0.6)
plt.xlabel('Medical Costs')
plt.ylabel('Number of Patients')
plt.legend(['Smoker','Non-Smoker'])
plt.title('Distribution of Medical Charges')
plt.show()
```

```python
older = data.loc[data['age'] > 40, 'charges']
younger = data.loc[data['age'] < 40, 'charges']

plt.hist(older,15)
plt.hist(younger,15,alpha=0.7)
plt.xlabel('Medical Costs')
plt.ylabel('Number of Patients')
plt.legend(['Older','Younger'])
plt.title('Distribution of Medical Charges')
plt.show()
```

```python
male = data.loc[data['sex'] == 1, 'charges']
female = data.loc[data['sex'] == 0, 'charges']

plt.hist(male,12)
plt.hist(female,12,alpha=0.6)
plt.xlabel('Medical Costs')
plt.ylabel('Number of Patients')
plt.legend(['Male','Female'])
plt.title('Distribution of Medical Charges')
plt.show()
```

```python
southwest = data.loc[data['region'] == 0, 'charges']
southeast = data.loc[data['region'] == 1, 'charges']
northwest = data.loc[data['region'] == 2, 'charges']
northeast = data.loc[data['region'] == 3, 'charges']

regions = [southwest,southeast,northwest,northeast]
labels = list(['Southwest','Southeast','Nortwest','Northeast'])
fig, ax = plt.subplots()
ax.boxplot(regions,0, '',labels=labels)

plt.title('Medical Charges by Region')
plt.show()
```
